```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# خواندن ویدیو
video = cv2.VideoCapture('Indiantraffic.avi')

# نمایش یک فریم دلخواه
ret, frame = video.read()
if ret:
    # تبدیل به سیاه و سفید
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # افزودن نویز Salt-and-pepper
    noise_frame = np.copy(gray_frame)
    noise_prob = 0.02
    black_noise = np.random.rand(*gray_frame.shape) < noise_prob
    white_noise = np.random.rand(*gray_frame.shape) > 1 - noise_prob
    noise_frame[black_noise] = 0
    noise_frame[white_noise] = 255

    # نمایش فریم با نویز
    plt.imshow(noise_frame, cmap='gray')
    plt.title('Frame with Salt-and-Pepper Noise')
    plt.show()

    # چاپ رزولوشن تصویر
    print("Resolution:", gray_frame.shape)
else:
    print("Error in reading video")

# بستن فایل ویدیو
video.release()

```

```python
def add_salt_and_pepper_noise(image, noise_level=0.02):
    """
    افزودن نویز Salt-and-Pepper به تصویر
    :param image: تصویر ورودی
    :param noise_level: سطح نویز
    :return: تصویر با نویز
    """
    noisy_image = np.copy(image)
    num_salt = np.ceil(noise_level * image.size * 0.5).astype(int)
    num_pepper = np.ceil(noise_level * image.size * 0.5).astype(int)

    # اضافه کردن نویز نمک
    coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape]
    noisy_image[tuple(coords)] = 1

    # اضافه کردن نویز فلفل
    coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape]
    noisy_image[tuple(coords)] = 0

    return noisy_image


def convert_video_to_tensor_with_noise(video_path):
    video = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = video.read()
        if not ret:
            break
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        noisy_frame = add_salt_and_pepper_noise(gray_frame)
        frames.append(noisy_frame)

    video.release()
    video_tensor = np.stack(frames, axis=0)
    return video_tensor

video_path = 'Indiantraffic.avi'
video_tensor = convert_video_to_tensor_with_noise(video_path)
```

```python
import cv2
import numpy as np
from sklearn.utils.extmath import randomized_svd

def frobenius_norm(tensor):
    return np.linalg.norm(tensor)

def apply_randomized_svd(video_tensor, n_components=150):  # افزایش تعداد مؤلفه‌ها
    nsamples, nx, ny = video_tensor.shape
    d2_tensor = video_tensor.reshape((nsamples, nx * ny))

    U, Sigma, VT = randomized_svd(d2_tensor, n_components=n_components)
    L = np.dot(U, np.dot(np.diag(Sigma), VT))
    S = video_tensor - L.reshape(video_tensor.shape)
    reconstruction_error = frobenius_norm(video_tensor - L.reshape(video_tensor.shape))
    return L.reshape(video_tensor.shape), S, reconstruction_error

video_path = 'Indiantraffic.avi'
video_tensor = convert_video_to_tensor_with_noise(video_path)
L, S, reconstruction_error = apply_randomized_svd(video_tensor)
print("Reconstruction Error:", reconstruction_error)
```

```python
import cv2
import matplotlib.pyplot as plt

# انتخاب یک فریم دلخواه
frame_number = 100  # به عنوان مثال فریم شماره 100

# فریم از تنسور Low Rank
low_rank_frame = L[frame_number]
# فریم از تنسور Sparse
sparse_frame = S[frame_number]

# نمایش فریم‌ها
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(low_rank_frame, cmap='gray')
plt.title('Low Rank Frame')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sparse_frame, cmap='gray')
plt.title('Sparse Frame')
plt.axis('off')

plt.show()
```

```python
def create_video_from_frames(frames, output_path, fps=20):
    height, width = frames[0].shape
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), fps, (width, height), isColor=True)

    for frame in frames:
        # تبدیل فریم به رنگی (BGR)
        color_frame = cv2.cvtColor(frame.astype('uint8'), cv2.COLOR_GRAY2BGR)
        out.write(color_frame)

    out.release()

# ایجاد ویدیوی 'foreground.avi'
create_video_from_frames(S, 'foreground.avi')

# ایجاد ویدیوی 'background.avi'
create_video_from_frames(L, 'background.avi')

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری ویدیو
cap = cv2.VideoCapture('background.avi')

# اطمینان از باز شدن ویدیو
if not cap.isOpened():
    raise IOError("Cannot open video")

# دریافت ابعاد فریم‌های ویدیو
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# بارگذاری و تغییر اندازه تصویر نهان‌شده
watermark_image = cv2.imread('1.JPG', cv2.IMREAD_GRAYSCALE)
watermark_image_resized = cv2.resize(watermark_image, (frame_width, frame_height))

# تغییر دادن مقیاس تصویر نهان‌شده
watermark_image_resized = watermark_image_resized / 255.0  # نرمال‌سازی به محدوده [0, 1]

# تابعی برای نهان‌سازی ویدیو
def embed_watermark(frame, watermark):
    U, S, Vt = np.linalg.svd(frame, full_matrices=False)
    S_mod = S + watermark.flatten()[:len(S)]
    watermarked_frame = U @ np.diag(S_mod) @ Vt
    watermarked_frame_normalized = cv2.normalize(watermarked_frame, None, 0, 255, cv2.NORM_MINMAX)
    watermarked_frame_uint8 = watermarked_frame_normalized.astype('uint8')
    return watermarked_frame_uint8

# تابعی برای اعمال فیلتر Blur
def apply_blur(frame):
    return cv2.GaussianBlur(frame, (5, 5), 0)

# تابعی برای ذخیره‌سازی ویدیو
def save_video(frames, path):
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(path, fourcc, 30.0, (frame_width, frame_height), True)
    for frame in frames:
        out.write(frame)
    out.release()

# پردازش فریم‌های ویدیو و نهان‌سازی واترمارک
watermarked_frames = []
blurred_frames = []
ret, frame = cap.read()
if ret:
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    watermarked_frame = embed_watermark(gray_frame, watermark_image_resized)
    blurred_frame = apply_blur(watermarked_frame)
    watermarked_frames.append(watermarked_frame)
    blurred_frames.append(blurred_frame)

# نمایش فریم قبل و بعد از اعمال واترمارک و فیلتر Blur
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(watermarked_frames[0], cmap='gray')
plt.title('Watermarked Frame')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred_frames[0], cmap='gray')
plt.title('Blurred Watermarked Frame')
plt.axis('off')

plt.show()

# ذخیره ویدیوی نهایی
save_video(blurred_frames, 'blured-background-watermarked.avi')

# رسم نمودار مقادیر تکین قبل و بعد از نهان‌سازی
U, S, Vt = np.linalg.svd(gray_frame, full_matrices=False)
U_w, S_w, Vt_w = np.linalg.svd(watermarked_frames[0], full_matrices=False)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(S, label='Original')
plt.title('Singular Values (Original)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(S_w, label='Watermarked')
plt.title('Singular Values (Watermarked)')
plt.legend()

plt.show()


# بستن فایل‌ها
cap.release()
```
